### Class Weights

```python
from pprint import pprint as pp
```

```python
def create_class_weights(y, smooth_factor=0.2):
        counter = Counter(y)
        if smooth_factor > 0:
            p = max(counter.values()) * smooth_factor
            for k in counter.keys():
                counter[k] += p
        majority = max(counter.values())
        return {cls: float(majority / count) for cls, count in counter.items()}
```

```python
weights = create_class_weights(train_generator.classes.astype(int), 0.2)
pp(weights)
```

```python
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    plt.figure(figsize = (12,12))
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(num_classes)
    plt.xticks(tick_marks, classes, rotation=90)
    plt.yticks(tick_marks, classes)
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")
    #plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

### Unfreeze the resnet layers for training

```python
model.load_weights('first_trans_1.h5')
```

```python
# define the parameters in the same way as above, set which layers we will now unlock and how many epochs to train with the new settings

split_at = 10

for layer in base_model.layers[:split_at]: 
   layer.trainable = False

for layer in base_model.layers:
    layer.trainable = True

from keras.optimizers import Adam

optimizer_2 = Adam(lr=0.005, beta_1=0.9, beta_2=0.999, epsilon=1e-08, decay=0.0)    

model.compile(optimizer=optimizer_2, loss='categorical_crossentropy', metrics=['accuracy', 'mae'])

model.summary()
```

```python
from keras.callbacks import ModelCheckpoint, LearningRateScheduler, EarlyStopping, ReduceLROnPlateau
# set the model to stop early automatically if results are not improving
early = EarlyStopping(monitor="val_loss", 
                      mode="min", 
                      patience=3)
```

```python
%%time

num_epochs = 30

# fit the model with the new parameters and continue training
history_2 = model.fit_generator( train_generator,
                                epochs=num_epochs,
                                class_weight=weights,
                                steps_per_epoch=(train_size // batch_size) + 1,
                                validation_data=validation_generator,
                                validation_steps=(validation_size // batch_size) + 1,
                                callbacks=[checkpoint, tensorboard, early])
```
